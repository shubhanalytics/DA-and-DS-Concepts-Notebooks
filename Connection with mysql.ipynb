## What if password already has a special characters, then we can use urllib.parse.quote_plus:


```python
from urllib.parse import quote_plus
from sqlalchemy import create_engine

password = quote_plus("AlphaMysql@999")  # Encodes special characters
engine = create_engine(f"mysql+pymysql://root:{password}@localhost/orders")

query = 'select * from order_items'
```

```python
order_items = pd.read_sql(query, engine)

order_items.head()
```
