```python
#First, let us see how to view the values of a specific column through the column names
#Let us call out the top 5 values of column name 'macdonalds_food_R'
data["macdonalds_food_R"].head()
# .head() - return the top 5 values
```

```python
# Now let us see how to view the food ratings together for all 4 columns.
# As these are multiple values, you have to first pass a list (names of all 4 columns) into the 'getitem' syntax (the []'s)

data[["macdonalds_food_R","KFCT_food_R","Sagir_Ratna_food_R","Borger_King_Food_R"]].head()
# .head() - return the top 5 values

```

Now, we know how to seperate the food ratings and service ratings from the data. Let us check out the mean and median ratings for both

```python
data[["macdonalds_food_R","KFCT_food_R","Sagir_Ratna_food_R","Borger_King_Food_R"]].mean()
```

```python
data[["macdonalds_food_R","KFCT_food_R","Sagir_Ratna_food_R","Borger_King_Food_R"]].median()
```

Mean and Median are both measures of central tendancy. One common measure of central tendancy that we use in our day to day life is average. To know more about both, refer to the video by Dr Abhinanda Sarkar
The above 2 output shows that 
#### People rate the food of Borger King much higher in comparison to the other 3 restaurents of the institute. 

Let us also look at the spread of food ratings for all 4 restaurents. Standard Deviation is one such measure that tells us about the variation in the data. To know more about Standard Deviation, refer to the video by Dr Abhinanda Sarkar.

```python
data[["macdonalds_food_R","KFCT_food_R","Sagir_Ratna_food_R","Borger_King_Food_R"]].std()
```

This tells you that there is more variation in food ratings for Sagir Ratna restaurent. So they require to be more consistent with the quality of the food.

A report will be useless without some visualization. Let us first pick the highest rated restaurent in terms of food and explore it

#### Histogram for Burger_king_food_R

```python
data["Borger_King_Food_R"].hist() 

# .hist() - Returns a nice histogram of the distribution
## . A histogram is a graph that organizes and displays numerical data in picture form,
## showing groups of data and the number or percentage of the data that fall into each group. It gives you a nice snapshot
## of the data set.. Here we can see the frequency for Ratings are on the y-axis and Ratings are on the x-axis.
```

Looking at the histogram we can see that for rating =7,it is showing the highest frequency.
#### Most people have rated the food as 7 for Borger king.

#### Pie-Chart for Burger_king_food_R

```python
data["Borger_King_Food_R"].value_counts()
## The value_counts() function is used to get a Series containing counts of unique values.
## The resulting object will be in descending order so that the first element is the most frequently-occurring element.
```

```python
data["Borger_King_Food_R"].value_counts().plot()
```

I know, the above graph doesn't look so visually appealing. A pie chart would be more useful in this case. 

```python
data["Borger_King_Food_R"].value_counts().plot(kind="pie", autopct ='%.2f%%')
# The slices of a pie chart depicts the proportion of the ratings
# Try the same command without the autopct and observe the results
```

The autopct enables you to display the percent value using Python string formatting. For example, if autopct='%. 2f' , then for each pie wedge, the format string is '%. 2f' and the numerical percent value for that wedge is pct , so the wedge label is set to the string '%

#### Bar chart for Burger_king_food_R

```python
data["Borger_King_Food_R"].value_counts().plot(kind="bar") ## it is also another way to see the distribution 
```
